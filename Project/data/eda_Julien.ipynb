```python
import pandas as pd
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("~/Project/data/emissions_dataset.csv")
```

```python
columns_to_remove = ['Hard Coal Export', 'Natural Gas Export', "Crude Oil Export", 'Hard Coal Production' , 'Natural Gas Production' , "Crude Oil Production", "Unnamed: 0"]
df.drop(columns_to_remove, axis = 1, inplace = True)
df.head()
```

```python
# Show correlation matrix
fig = px.imshow(df.corr(),title="Correlation Matrix",width=650,height=650)
fig.update_xaxes(tickangle=-45)
```

```python
#gas oil/deisel oil v. Motor Gasoline Prod
#gas oil/deisel oil & Motor Gasoline v. Green house gas emissions
```

```python
fig = px.line(df, x = "Year", y = "Greenhouse Gas Emissions", color = "Country", title = "Greenhouse Gas Emissions v. Time")
fig.show()
fig.write_html('./Charts/GHG_time.html')
```

```python
fig = px.pie(df, value = "Greenhouse Gas Emissions", head = "country", title = "Share of Greenhouse Gas emissions")
fig.show()
```
