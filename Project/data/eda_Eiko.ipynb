```python
import pandas as pd
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("~/Project/data/emissions_clean.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
len(df)
```

```python
columns_to_remove = ['Hard Coal Export', 'Natural Gas Export', "Crude Oil Export", 'Hard Coal Production' , 'Natural Gas Production' , "Crude Oil Production", "Unnamed: 0"]
df.drop(columns_to_remove, axis = 1, inplace = True)
df.head()
```

```python
# Show correlation matrix
fig = px.imshow(df.corr(),title="Correlation Matrix",width=650,height=650)
fig.update_xaxes(tickangle=-45)
```

```python
fig = px.line(df, x = "Year", y = "Greenhouse Gas Emissions", color = "Country", title = "Greenhouse Gas Emissions v. Time")
fig.show()
```

```python
fig = px.scatter(df[df.Country == "RUS"], x = "Year", y = "Greenhouse Gas Emissions")
fig.show()
```

```python
fig = px.pie(df, values = "Natural Gas Import", names = "Country", title= 'Natural Gas Imports by country')
fig.show()
```

```python
fig = px.pie(df, values = "Natural Gas Export", names = "Country", title= 'Natural Gas Exports by country')
fig.show()
```
