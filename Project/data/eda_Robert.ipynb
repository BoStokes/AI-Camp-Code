```python
import pandas as pd
import plotly.express as pl
```

```python
df = pd.read_csv("emissions_dataset.csv")
df
```

```python
MAX_MISSING = 5
df.drop(df[df.isnull().sum(axis=1)>MAX_MISSING].index,axis=0,inplace=True)
df.reset_index(drop=True, inplace=True)
df.info()
```

```python
df["Country"].unique()
```

```python
fig = pl.scatter(df, x="Year", y="Natural Gas Production", size="Greenhouse Gas Emissions", color = "Country", title="Natural Gas Production of Different Countries Over Time")
```

```python
fig = pl.scatter(df, x="Year", y="Crude Oil Production", size="Greenhouse Gas Emissions", color = "Country", title="Crude Oil Production of Different Countries Over Time")
fig.write_html('./Charts/Crude Oil Production of Different Countries Over Time.html')
fig
```

```python
fig = pl.scatter(df, x="Year", y="Hard Coal Production", size="Greenhouse Gas Emissions", color = "Country", title="Hard Coal Production of Different Countries Over Time")
fig.write_html('./Charts/Hard Coal Production of Different Countries Over Time.html')
fig
```
