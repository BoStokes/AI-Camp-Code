```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("../../Project/data/emissions_dataset.csv")
```

```python
df.columns.values
```

```python
fig = px.scatter(df, x="Gas Oil/Diesel Oil Import", y="Greenhouse Gas Emissions", color='Country', title='Gas Oil/Diesel Oil Import vs. Greenhouse Gas Emissions')
fig.write_html('./Charts/Gas Oil Diesel Oil Import vs. Greenhouse Gas Emissions.html')
fig
```

```python
fig = px.line(df, x="Year", y="Hard Coal Production", color='Country', title='Hard coal production over time')
fig.show()
```

```python
only_large = df[df['Greenhouse Gas Emissions'] > 10000000]
fig = px.pie(only_large, values='Greenhouse Gas Emissions', names='Country', title='Greenhouse gas emissions by country', height=800)
fig.show()
```
